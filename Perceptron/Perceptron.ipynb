```python
import pandas as pd
import numpy as np
```

```python
def perceptronStandard(trainData,testData,T=10,r=1):
    n = len(trainData.columns)
    m = len(trainData.index)
    ones = np.ones(m)
    w = np.zeros((1,n))
    for i in range(T):
        d = trainData.sample(frac=1).reset_index(drop=True).to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        for j in range(m):
            if(y[j]*np.dot(w,x[j]) <= 0):
                w = w + r * y[j] * x[j]
    print(w)
    m = len(testData.index)
    ones = np.ones(m)
    count = 0
    for j in range(m):
        d = testData.sample(frac=1).reset_index(drop=True).to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        if(y[j] * np.dot(w,x[j]) > 0):
            count += 1
    print(count/m)
```

```python
def perceptronVoted(trainData,testData,T=10,r=1):
    n = len(trainData.columns)
    m = len(trainData.index)
    ones = np.ones(m)
    w = np.zeros((T*m,n))
    c = np.zeros(T*m)
    k = 0
    for i in range(T):
        d = trainData.sample(frac=1).reset_index(drop=True).to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        for j in range(m):
            if(y[j]*np.dot(w[k],x[j]) <= 0):
                w[k+1] = w[k] + r * y[j] * x[j]
                k += 1
                c[k] = 1
            else:
                c[k] += 1
    for i in range(1,k+1):
        print(i,"&",w[i],"&",c[i],"\\\\")
    m = len(testData.index)
    ones = np.ones(m)
    count = 0
    for j in range(m):
        d = testData.to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        pred = 0
        for i in range(k+1):
            pred += c[i] * np.dot(w[i],x[j])
        if(y[j]*pred > 0):
            count += 1
    print(count/m)
```

```python
def perceptronAverage(trainData,testData,T=10,r=1):
    n = len(trainData.columns)
    m = len(trainData.index)
    ones = np.ones(m)
    w = np.zeros((1,n))
    a = np.zeros((1,n))
    for i in range(T):
        d = trainData.sample(frac=1).reset_index(drop=True).to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        for j in range(m):
            if(y[j]*np.dot(w,x[j]) <= 0):
                w = w + r * y[j] * x[j]
            a = a + w
    print(a)
    m = len(testData.index)
    ones = np.ones(m)
    count = 0
    for j in range(m):
        d = testData.sample(frac=1).reset_index(drop=True).to_numpy()
        x = d[:,:-1]
        x = np.insert(x,0,ones,axis=1)
        y = d[:,-1]
        if(y[j] * np.dot(a,x[j]) > 0):
            count += 1
    print(count/m)
```

```python
trainData = pd.read_csv("bank-note/train.csv",header=None)
testData = pd.read_csv("bank-note/test.csv",header=None)
```

```python
perceptronStandard(trainData,testData)
```

```python
perceptronVoted(trainData,testData)
```

```python
perceptronAverage(trainData,testData)
```
